```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# GY1 Star Test
# Set up file path and call the fils using pandas
gy1path = "/Users/cwray/Desktop/ASTR502/ASTR502-SP26/Wk19th/bhac15_t_1.dat"
with open(gy1path, 'r') as file:
    GY1content = pd.read_csv(gy1path, sep='\s+', comment='#', names=['t', 'm', 'teff', 'logg', 'logL', 'Lum', 'R'])
# Using sep='\s+' to seaparate terms by whitespaces rather than commas
# Uses comment='#' to ignore the titles when it comes to referencing data
```

```python
# Going to plot Teff (Effective Temperature) vs logL (Luminosity)
plt.figure(figsize=(8, 6))
plt.plot(GY1content['teff'], GY1content['logL'], marker='o', linestyle='-', color='darkred')

# Labeling the axes
plt.xlabel('Effective Temperature $Teff$ (K)')
plt.ylabel('Log Luminosity $\log(L/L_{\odot})$')
# Using _ can create subscripts
plt.title('1 Gyr Isochrone Plot Test')

plt.grid(True, linestyle='--', alpha=0.6)
```

```python
# Repeat setting up the file calling for Gy5 Star
gy5path = "/Users/cwray/Desktop/ASTR502/ASTR502-SP26/Wk19th/bhac15_t_5.dat"
with open(gy5path, 'r') as file:
    GY5content = pd.read_csv(gy5path, sep='\s+', comment='#', names=['t', 'm', 'teff', 'logg', 'logL', 'Lum', 'R'])
# Using sep='\s+' to seaparate terms by whitespaces rather than commas
# Uses comment='#' to ignore the titles when it comes to referencing data
```

```python
# Add Gy5 data to plot
plt.figure(figsize=(8, 6))
plt.plot(GY1content['teff'], GY1content['logL'], marker='o', linestyle='-', color='darkred', label = '1 Gyr')
plt.plot(GY5content['teff'], GY5content['logL'], marker='o', linestyle='-', color='darkblue', label = '5 Gyr')

# Labeling the axes
plt.xlabel('Effective Temperature $Teff$ (K)')
plt.ylabel('Log Luminosity $\log(L/L_{\odot})$')
# Using _ can create subscripts
plt.title('1 Gyr vs 5 Gyr Isochrone Plot (BHAC15)')
plt.legend(loc='upper left')

plt.grid(True, linestyle='--', alpha=0.6)
```
