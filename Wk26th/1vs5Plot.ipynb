```python
import os
os.environ["PYTHONIOENCODING"] = "utf-8"

import ezpadova
import matplotlib.pyplot as plt
```

```python
Gyr1 = ezpadova.get_one_isochrone(age_yr=1e9, Z=0.0152, label='parsec_v12s')
Gyr5 = ezpadova.get_one_isochrone(age_yr=5e9, Z=0.0152, label='parsec_v12s')
# Use get_isochrones() for a range of metallicities or ages
```

```python
# Define terms for plotting
gyr1logL = Gyr1[Gyr1['logL'] > -8]
gyr1logTe = Gyr1[Gyr1['logL'] > -8]

# df[df['Vmag'] > -5]

gyr5logL = Gyr5['logL']
gyr5logTe = Gyr5['logTe']
```

```python
#Plotting

plt.scatter(gyr1logTe, gyr1logL, color='firebrick', label='1 Gyr PARSEC')
plt.scatter(gyr5logTe, gyr5logL, color='navy', label='5 Gyr PARSEC')

plt.gca().invert_yaxis()  # Magnitudes are brighter at the top
plt.xlabel('log(Effective Temperature)', fontsize=12)
plt.ylabel('log(Luminosity)', fontsize=12)
plt.title('PARSEC Isochrone: 1 Gyr ($log_{10}t = 9$) vs 5 Gyr ($log_{10}t = 9.7$)', fontsize=14)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()

plt.show()
```
