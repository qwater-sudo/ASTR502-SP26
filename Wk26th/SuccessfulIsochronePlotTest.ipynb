```python
import os
os.environ["PYTHONIOENCODING"] = "utf-8"

import ezpadova
import matplotlib.pyplot as plt

# 1. Fetch data (Run this if 'df' isn't in memory yet)
# Note: Since your previous output showed Bmag/Vmag, we use the default system
df = ezpadova.get_one_isochrone(logage=9.0, Z=0.0152, label='parsec_v12s')

# 2. Plotting
plt.figure(figsize=(8, 10))

# Using the columns confirmed in your 'Available columns' list
color = df['Bmag'] - df['Vmag']
magnitude = df['Vmag']

plt.scatter(color, magnitude, color='firebrick', lw=2, label='1 Gyr PARSEC')

# Formatting
plt.gca().invert_yaxis() 
plt.xlabel('B - V (Color Index)', fontsize=12)
plt.ylabel('$M_V$ (Absolute Magnitude)', fontsize=12)
plt.title('PARSEC Isochrone: 1 Gyr ($log_{10}t = 9$)', fontsize=14)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()

plt.show()
```
