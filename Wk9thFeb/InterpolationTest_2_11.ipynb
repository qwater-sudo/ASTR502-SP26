```python
# Import necassary libraries and functions
import os
os.environ["PYTHONIOENCODING"] = "utf-8"

import ezpadova
from ezpadova import parsec
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# Using QuickIsochrone, requires either a specific PARSEC file or a pandas dataframe
# To get a pandas datframe from get_isochrones, set return_df = True

df = ezpadova.get_isochrones(logage=(8.9, 9.1, 0.05), MH=(0, 0, 0.1), return_df=True)
# The tuples for logage and MH are in format (min, max, step)
```

```python
print(df.columns.tolist())
```

```python

from scipy.interpolate import LinearNDInterpolator

# Choose your dimensions based on what actually changes in 'df'
points = df[['logAge', 'Mini']].values
values = df['logL'].values

# Create the Interpolator
interp_logL = LinearNDInterpolator(points, values)

# Call it with target age and mass
target_age = 9.02
target_mass = 1.15
result = interp_logL(target_age, target_mass)

print(f"Interpolated LogL: {float(result):.4f}")
```
